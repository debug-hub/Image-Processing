#### 2) Image Resizing

```python
g1 = cv2.imread('Images/apple.jpg')
g2 = cv2.cvtColor(g1,cv2.COLOR_BGR2RGB)
print(g1.shape)
plt.imshow(g2)
plt.show()
```

Interpolation Methods
1) cv2.INTER_AREA: This is used when we need to shrink an image.<br>
2) cv2.INTER_CUBIC: This is slow but more efficient.<br>
3) cv2.INTER_LINEAR: This is primarily used when zooming is required. This is the default interpolation technique in OpenCV.

```python
g3 = cv2.resize(g1,(400,400),interpolation=cv2.INTER_AREA)
g4 = cv2.resize(g1,(650,650),interpolation=cv2.INTER_LINEAR)
g5 = cv2.resize(g1,(900,600),interpolation=cv2.INTER_LINEAR)
cv2.imshow('original',g1)
cv2.imshow('small',g3)
cv2.imshow('large',g4)
cv2.imshow('disturbed',g5)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
g5 = cv2.resize(g1,(900,600),interpolation=cv2.INTER_LINEAR)
print(g5.shape)
cv2.imshow('Horizontally Stretched Image',g5)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
g6 = cv2.resize(g1,(700,1000),interpolation=cv2.INTER_CUBIC)
print('Vetically Stretched Image')
print(g6.shape)
cv2.imshow('Vertically Stretched Image',g5)
cv2.waitKey()
cv2.destroyAllWindows()
```

### Image Thresholding
Converts grayscale image to its binary form

```python
s1 = cv2.imread('Images/kneau.jpg',0)
print(s1.shape)
plt.imshow(s1,'gray')
plt.show()
```

```python
s1[30]
```

```python
# pixel values below 150 turns 0(black) and above 150 turns 255(white)
_,t1 = cv2.threshold(s1,150,255,cv2.THRESH_BINARY)
cv2.imshow('Original',s1)
cv2.imshow('Thresh Binary',t1)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
print(t1.shape)
```

```python
t1[30]
```

```python

_,t2 = cv2.threshold(s1,150,255,cv2.THRESH_BINARY_INV)
# print(r)
cv2.imshow('Original',s1)
cv2.imshow('Thresh Binary Inverse',t2)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
t2[30]
```

```python
# Pixel values above 150 are truncated (held at 150) and remaining pixels remain as it is
_,t3 = cv2.threshold(s1,150,255,cv2.THRESH_TRUNC)
# print(r)
cv2.imshow('Original',s1)
cv2.imshow('Thresh Binary Truncated',t3)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
t3[30]
```

```python
# Pixel values less than 150 are made 0 (black) and pixel values above 150 are held as it is
_,t4 = cv2.threshold(s1,150,255,cv2.THRESH_TOZERO)
# print(r)
cv2.imshow('Original',s1)
cv2.imshow('Thresh Binary To Zero',t4)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
t4[30]
```

```python
# pizel vales above 150 are turned into 0 and pixel values less than 150 are held as it is
_,t5 = cv2.threshold(s1,150,255,cv2.THRESH_TOZERO_INV)
cv2.imshow('Original',s1)
cv2.imshow('Thresh Binary ToZero Inv',t5)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
t5[30]
```

### Image Denoising

```python
nimg = cv2.imread('Images/noisy_img.jpg')
nimg = cv2.cvtColor(nimg,cv2.COLOR_BGR2RGB)
plt.imshow(nimg)
plt.show()
```

```python
k1 = np.ones((5,5))/25
blur2 = cv2.filter2D(nimg,-1,k1)
plt.imshow(blur2)
plt.show()
```

```python
k1 = np.ones((5,5))/25
b1 = cv2.filter2D(nimg,-1,k1)
b2 = cv2.blur(nimg,(5,5))
b3 = cv2.GaussianBlur(nimg,(5,5),0)
b4 = cv2.medianBlur(nimg,5)

plt.imshow(nimg),plt.title('Original'),plt.show()
plt.imshow(b1),plt.title('Filter2D'),plt.show()
plt.imshow(b2),plt.title('blur'),plt.show()
plt.imshow(b3),plt.title('GaussianBlur'),plt.show()
plt.imshow(b4),plt.title('MedianBlur'),plt.show()
plt.show()
```

```python
nimg2 = cv2.imread('Images/e1.jpg')
plt.imshow(nimg2)
plt.show()
```

```python
k1 = np.ones((5,5))/25
b1 = cv2.filter2D(nimg2,-1,k1)
b2 = cv2.blur(nimg2,(5,5))
b3 = cv2.GaussianBlur(nimg2,(5,5),0)
b4 = cv2.medianBlur(nimg2,5)
b5 = cv2.bilateralFilter(nimg2,9,75,75)

plt.imshow(nimg2),plt.title('Original'),plt.show()
plt.imshow(b1),plt.title('Filter2D'),plt.show()
plt.imshow(b2),plt.title('blur'),plt.show()
plt.imshow(b3),plt.title('GaussianBlur'),plt.show()
plt.imshow(b4),plt.title('MedianBlur'),plt.show()
plt.imshow(b5),plt.title('Bilateral Filtering'),plt.show()
plt.show()
```

### Image Sharpening

```python
q1 = cv2.imread('Images/scene1.jpg')
# q1 = cv2.cvtColor(q1,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(7,7))
plt.imshow(q1)
plt.show()
```

```python
k1 = np.array([[-1,-1,-1],[-1,9,1],[-1,-1,-1]]) # Sharpening
k2 = np.array([[0,-1,-1],[1,0,-1],[1,1,1]])      # Emboss

s1 = cv2.filter2D(q1,-1,k1)
s2 = cv2.filter2D(q1,-1,k2)

plt.imshow(q1),plt.title('Original'),plt.show()
plt.imshow(s1),plt.title('k1 Filter'),plt.show()
plt.imshow(s2),plt.title('k2 filter'),plt.show()
```
